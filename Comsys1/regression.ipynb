```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

**Here we've used normal LinearRegression to predict values without using a train_test_split**

```python
#Getting the train.csv dataset
url='https://drive.google.com/file/d/12QJ_ZtwhAGQ346n3UbSta_fn1cNMoIEv/view?usp=drive_link'
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
train_data= pd.read_csv(dwn_url)

train_data.shape
```

```python
train_data.head(2)
```

```python
#Dropping the Name and Country columns because they won't help us in predictions 
train_data=train_data.drop("Name",axis=1)
train_data=train_data.drop("Country",axis=1)
```

```python
#Checking for NaN type values
train_data.isnull().sum().sum()
```

```python
#Filling the NaN type values with 0
df = train_data.fillna(value=0)
```

```python
#Using the entire data for training instead of applying a train_test_split because on applying train_test_split the private score was evaluating to a lesser value than what we got on this try
X_train = df.drop("Value at beginning of 2023/24 season", axis=1)
y_train =df["Value at beginning of 2023/24 season"]
```

```python
#Training the Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
#Getting the test.csv dataset
url='https://drive.google.com/file/d/1DSJh_G5ERRbHCG2gY_P1BHS47p6qsdmU/view?usp=drive_link'
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
df2 = pd.read_csv(dwn_url)

df2
```

```python
df2=df2.drop("Country",axis=1)
```

```python
df2.head(2)
```

```python
#Checking for NaN type values
df2.isnull().sum().sum()
```

```python
#Filling NaN type values with 0
test_data = df2.fillna(value =0)
```

```python
#Predicting values in a numpy array
predictions = linear_model.predict(test_data)
```

```python
predictions
```

```python
#Adding the numpy values to a pandas dataframe
df_output = pd.DataFrame(predictions)
```
